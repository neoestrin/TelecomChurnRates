# Training the model

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import preprocessing

y = data["Churn"]
x = data
x.pop("Churn")

norm_x = preprocessing.normalize(x)

trainx, testx, trainy, testy = train_test_split(norm_x, y, test_size = 0.3)      
```

# Logistic Regression

```
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import metrics

logreg = LogisticRegression()
logreg.fit(trainx, trainy)

y_pred = logreg.predict(testx)
print(np.sqrt(mean_squared_error(testy,y_pred))) 
score = logreg.score(testx, testy)
print(score)

cm = metrics.confusion_matrix(testy, y_pred)
print(cm)
```

The RMSE value is 0.31. This means that the average difference from the best fit line for the Churn is 0.31.

And our accuracy in this model is 90.25%
